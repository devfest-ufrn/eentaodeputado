### Nosso dataset *deputados.csv* não possui o campo de Id cadastro do deputado(nossa "conexão" com as propostas)

Vamos tentar fazer o fetch diretamente da api de dados abertos da câmara dos deputados

```python
import requests as re
import xmltodict as xml2dict
from collections import OrderedDict
```

```python
response = re.get('http://www.camara.leg.br/SitCamaraWS/Deputados.asmx/ObterDeputados')
```

```python
tree = xml2dict.parse(response.content)
```

```python
tree.keys()
```

```python
deputados = tree['deputados']['deputado']
```

```python
deputados
```

### Voilà! :D

```python
deputados_df = pd.DataFrame(deputados)
deputados_df.head()
```
