```python
import pandas as pd
import numpy as np
import csv
```

```python
url_df = '../data/deputados-alterado.csv'
df = pd.read_csv(url_df)
```

```python
for index, row in df.iterrows():
    df.loc[index, 'produtividade'] = (((row['qtdPropostas']/df['qtdPropostas'].max()*3.0)
                                          + ((row['qtdPresencas']/df['qtdPresencas'].max())*6.0)
                                          - ((row['qtdAusencias']/df['qtdAusencias'].max())*1.0)))
```
